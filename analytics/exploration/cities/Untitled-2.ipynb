```python
import json
import pandas as pd
from shapely.geometry import shape, Point

# Charger le fichier GeoJSON
with open("/Users/raphaelcourivaud/Downloads/operations-programmees-open-data.json") as f:
    geojson_data = json.load(f)

# Charger le fichier CSV
csv_file = "../../dbt/seeds/cities.csv"  # Remplacez par le chemin de votre fichier CSV
csv_data = pd.read_csv(csv_file)

# Créer une liste pour stocker les résultats
results = []

# Parcourir chaque géométrie du GeoJSON
for feature in tqdm(geojson_data['features']):
    polygon = shape(feature['geometry'])
    geojson_name = feature['properties'].get('name', 'unknown')  # Remplacez 'name' par la clé correcte
    properties = feature["properties"]

    # Vérifier chaque point du CSV pour voir s'il est dans le polygone
    for _, row in csv_data.iterrows():
        point = Point(row['longitude'], row['latitude'])
        try:
            if polygon.contains(point):
                # Ajouter la ligne au DataFrame si le point est dans le polygone
                results.append({
                    'insee_code': row['insee_code'],
                    'city_code': row['city_code'],
                    'zip_code': row['zip_code'],
                    'label': row['label'],
                    'latitude': row['latitude'],
                    'longitude': row['longitude'],
                    'department_name': row['department_name'],
                    'department_number': row['department_number'],
                    'region_name': row['region_name'],
                    'code':properties['code'],
                    'nom':properties['nom'],
                    'typeprogramme':properties['typeprogramme'],
                    'datesignature':properties['datesignature']
                })
        except: 
            print(geojson_name)

# Convertir les résultats en DataFrame
results_df = pd.DataFrame(results)

# Afficher le DataFrame
results_df.head()

# Sauvegarder le DataFrame en CSV si nécessaire
results_df.to_csv('/path/to/output_file.csv', index=False)
```
