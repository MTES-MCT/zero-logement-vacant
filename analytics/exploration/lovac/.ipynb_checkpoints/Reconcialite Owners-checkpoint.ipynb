```python
from collections import defaultdict


ff_owners_columns_separated = defaultdict(list)
for col in ff_owners_columns:
    index = col.split(" ")[-1]
    ff_owners_columns_separated[index].append(col)
```

```python
import numpy as np


def clean_df_ff(df, index=index):
    address_cols = df.filter(regex="lign").columns
    name_col = df.filter(regex="denom").columns[0]
    date_col = df.filter(regex="datnss").columns[0]
    
    df[f"ff_owner_{index}_address"] = np.add.reduce(df[address_cols].fillna("") + " ", axis=1)
    df[f"ff_owner_{index}_name"] = df[name_col].str.replace("-", " ")
    df[f"ff_owner_{index}_birth_date"] = pd.to_datetime(df[date_col], errors="ignore", dayfirst=True)
    df = df.drop(list(address_cols) + [name_col, date_col], axis=1)
    return df
    
```

```python
from tqdm.auto import tqdm
```

```python
df_owners = pd.read_csv("data/zlovac_2023.csv", usecols=["Invariant"] + ff_owners_columns + biscom_owners_cols)

for index, cols in ff_owners_columns_separated.items():
    df_owners = clean_df_ff(df_owners, index=index)
```

```python
def _clean_m_mme_from_name(x: str):
    if isinstance(x, str):
        x = x.replace("-", " ")
        if x.startswith("M "): 
            return x[2:]
        elif x.startswith("MME "):
            return x[4:]
    return x
```

```python
def clean_df_biscom(df):
    address_cols = ["Adresse1", "Adresse2","Adresse3","Adresse4"]
    
    df["biscom_owner_address"] = np.add.reduce(df[address_cols].fillna("") + " ", axis=1)
    df["biscom_owner_name"] = df["Proprietaire"].fillna(df["Gestre Ppre"])
    df["biscom_owner_name"] = df["biscom_owner_name"].apply(_clean_m_mme_from_name)
    df = df.drop(list(address_cols) + ["Proprietaire", "Gestre Ppre"], axis=1)
    return df
```

```python
df_owners = clean_df_biscom(df_owners)
```

```python
for index in range(1,7):
    df_owners[f"match_extacly_{index}"] = df_owners[f'ff_owner_{index}_name'] == df_owners['biscom_owner_name']
    df_owners[f"match_ff_include_biscom_{index}"] = df_owners.apply(
        lambda x: str(x["biscom_owner_name"]) in str(x[f"ff_owner_{index}_name"]), axis=1 
    )
    df_owners[f"match_biscom_include_ff_{index}"] = df_owners.apply(
        lambda x: str(x[f"ff_owner_{index}_name"]) in str(x["biscom_owner_name"]), axis=1 
    )
```

## Handle First names

```python
df_prenom = pd.read_csv("./data/prenom.csv")
list_first_names = df_prenom[df_prenom["sum"] > 200]["prenom"].tolist()
```

```python
#!pip install flashtext
```

```python
from flashtext import KeywordProcessor
keyword_processor = KeywordProcessor()
keyword_processor.add_keywords_from_list(list_first_names)
```

```python
df_owners["biscom_owner_first_names"] = df_owners["biscom_owner_name"].fillna("").apply(keyword_processor.extract_keywords)
for index in range(1,7):
    df_owners[f"ff_owner_{index}_first_names"] = df_owners[f"ff_owner_{index}_name"].fillna("").apply(keyword_processor.extract_keywords)
    df_owners[f"match_same_first_name_{index}"] = df_owners.apply(
        lambda x: len(set(x["biscom_owner_first_names"]).intersection(set(x[f"ff_owner_{index}_first_names"]))) >= 1, axis=1)
```

## Fuzzy 

```python
#!pip install thefuzz
```

```python
from thefuzz import fuzz
FUZZY_RATIO = 75
```

```python
for index in range(1,7):
    df_owners[f"fuzzy_partial_ratio_{index}"] = df_owners.apply(
        lambda x: fuzz.partial_ratio(x["biscom_owner_name"],x[f"ff_owner_{index}_name"]), axis=1)

    df_owners[f"fuzzy_partial_token_sort_ratio_{index}"] = df_owners.apply(
        lambda x: fuzz.partial_token_sort_ratio(x["biscom_owner_name"],x[f"ff_owner_{index}_name"]), axis=1)

    df_owners[f"fuzzy_token_sort_ratio_{index}"] = df_owners.apply(
        lambda x: fuzz.token_sort_ratio(x["biscom_owner_name"],x[f"ff_owner_{index}_name"]), axis=1)

    df_owners[f"fuzzy_token_set_ratio_{index}"] = df_owners.apply(
        lambda x: fuzz.token_set_ratio(x["biscom_owner_name"],x[f"ff_owner_{index}_name"]), axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2)
df_owners["fuzzy_partial_ratio"].hist(ax=axes[0,0])
df_owners["fuzzy_partial_token_sort_ratio"].hist(ax=axes[0,1])
df_owners["fuzzy_token_sort_ratio"].hist(ax=axes[1,0])
df_owners["fuzzy_token_set_ratio"].hist(ax=axes[1,1])
```

```python
df_owners["fuzzy_ratio"] = df_owners[["fuzzy_partial_ratio", "fuzzy_token_set_ratio", "fuzzy_token_sort_ratio", "fuzzy_partial_token_sort_ratio"]].max(axis=1)
```

```python
df_owners["match_fuzzy"] = df_owners["fuzzy_ratio"] > FUZZY_RATIO
```
